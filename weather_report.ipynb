```python
data['Date_Time']=pd.to_datetime(data['Date_Time'])
```

```python
data
```

```python
import numpy as np
```

```python
data['month']=data['Date_Time'].dt.month
```

```python
monthly_avg_temp=data.groupby('month')['Temperature_C'].mean()
```

```python
monthly_avg_temp
```

```python
hottest_day=data.loc[data["Temperature_C"].idxmax()]
hottest_day
```

```python
coldest_day=data.loc[data["Temperature_C"].idxmin()]
coldest_day
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(monthly_avg_temp.index,monthly_avg_temp.values,marker='o', color='green')
plt.title("Monthly Average Temperature Trend")
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```

```python
monthly_avg_rainfall=data.groupby('month')['Precipitation_mm'].mean()
```

```python
plt.plot(monthly_avg_rainfall.index,monthly_avg_rainfall.values,marker='o', color='blue')
plt.title("Monthly Average Rainfall Trend")
plt.xlabel('Month')
plt.ylabel('Average Precipitation')
plt.grid()
plt.show()
```

```python
monthly_avg_humidity=data.groupby('month')['Humidity_pct'].mean()
```

```python
plt.plot(monthly_avg_humidity.index,monthly_avg_humidity.values,marker='o', color='red')
plt.title("Monthly Average Humidity Trend")
plt.xlabel('Month')
plt.ylabel('Average Humidity')
plt.grid()
plt.show()
```

```python
monthly_temp_varience=data.groupby('month')['Temperature_C'].var()
```

```python
monthly_temp_varience.plot(kind='bar',color='orange',figsize=(10,5))
plt.title('Monthly Temperature Varience')
plt.xlabel('Month')
plt.ylabel('Temperature Varience')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(data['Temperature_C'],kde=True,color='purple',bins=15)
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequncy')
plt.show()
```

```python
sns.histplot(data['Precipitation_mm'],kde=True,color='blue',bins=15)
plt.title('Rainfall Distribution')
plt.xlabel('Precipitation')
plt.ylabel('Frequncy')
plt.show()
```
